```python

import pandas as pd

df = pd.read_csv(
    "imoveis.csv",
    sep=";",
    encoding="latin1",
    na_values=["", "NULL", "null", "N/A"]
)

print(df.shape)
df.head()
# estrutura dos dados
df.info()

print(df.describe())

```

```python


# ===============================
# 2. Remover espaços extras
# ===============================

for col in df.select_dtypes(include="object").columns:
    df[col] = df[col].str.strip()
    
print("Espaços extras removidos com sucesso.")
print(df.describe())

```

```python
# ===============================
# 3. Padronizar texto
# ===============================

for col in df.select_dtypes(include="object").columns:
    df[col] = df[col].str.title()
print("Texto padronizado com sucesso.")
print(df.describe())

```

```python
# ===============================
# 4. Identificar dados faltantes
# ===============================

print(df.isnull().sum())

# remover linhas completamente vazias
df = df.dropna(how="all")
print("Linhas completamente vazias removidas com sucesso.")
print(df.describe())
```

```python

# ===============================
# 5. Converter colunas numéricas
# ===============================

for col in df.columns:
    df[col] = pd.to_numeric(df[col], errors="ignore")
print("Colunas numéricas convertidas com sucesso.")
print(df.describe())

```

```python
# ===============================
# 6. Remover duplicados
# ===============================

print("Duplicados:", df.duplicated().sum())

df = df.drop_duplicates()
print("Duplicados removidos com sucesso.")
print(df.describe())

```

```python
# ===============================
# 7. Estatísticas do dataset
# ===============================

print(df.describe())

```

```python
 #===============================
# 8. Salvar dataset limpo
# ===============================

df.to_csv("imoveis_limpo.csv", index=False)

print("Dataset limpo salvo com sucesso.")
```
