# Limpeza do Dataset imoveis.csv
Notebook para avaliação, identificação de problemas e tratamento dos dados.

```python
import pandas as pd
import numpy as np
```

```python
# Carregando dataset
file_path = "imoveis.csv"

try:
    df = pd.read_csv(file_path, sep=";", encoding="latin1")
except:
    df = pd.read_csv(file_path, sep=",", encoding="latin1")

df.head()
```

```python
# Análise inicial
print("Dimensões iniciais:", df.shape)
print("\nTipos de dados:\n", df.dtypes)
print("\nValores nulos:\n", df.isnull().sum())
```

```python
# Remover duplicados
df = df.drop_duplicates()

# Padronizar nomes das colunas
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

# Remover espaços extras
for col in df.select_dtypes(include="object").columns:
    df[col] = df[col].str.strip()

# Substituir valores vazios por NaN
df.replace(["", " ", "NA", "N/A", "null"], np.nan, inplace=True)

# Converter colunas numéricas possíveis
for col in df.columns:
    try:
        df[col] = pd.to_numeric(df[col])
    except:
        pass

# Remover registros com mais de 50% de nulos
df = df[df.isnull().mean(axis=1) < 0.5]

# Preencher nulos numéricos com mediana
for col in df.select_dtypes(include=["float64", "int64"]).columns:
    df[col] = df[col].fillna(df[col].median())

# Preencher nulos categóricos
for col in df.select_dtypes(include="object").columns:
    df[col] = df[col].fillna("Não informado")

df.head()
```

```python
# Salvando dataset limpo
df.to_csv("imoveis_limpo.csv", index=False)

print("Arquivo imoveis_limpo.csv gerado com sucesso!")
```
